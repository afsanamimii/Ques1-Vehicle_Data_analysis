# Pre Processing

```python
ex_df2=pd.json_normalize(json.loads(pd.DataFrame(df2['data']).to_json(orient='records')))
```

```python
ex_df2
```

```python
df_start2=ex_df2.loc[:,['data.time','data.Responses']]
```

```python
df_start2=df_start2.explode('data.Responses')
```

```python
df_start2
```

```python
ex_df2=pd.json_normalize(json.loads(pd.DataFrame(df_start2).to_json(orient='records')))
```

```python
ex_df2.drop('data.Responses', inplace=True, axis=1)
```

```python
ex_df2
```

```python
df_start2=ex_df2.loc[:,['data.time','data.Responses.lines']]
```

```python
df_start2=df_start2.explode('data.Responses.lines')
```

```python
ex_df2=pd.json_normalize(json.loads(pd.DataFrame(df_start2).to_json(orient='records')))
```

```python
ex_df2.drop('data.Responses.lines', inplace=True, axis=1)
```

```python
ex_df2
```

```python
df_start2=ex_df2.loc[:,['data.time','data.Responses.lines.lineId','data.Responses.lines.vehiclePositions']]
```

```python
df_start2=df_start2.explode('data.Responses.lines.vehiclePositions')
```

```python
ex_df2=pd.json_normalize(json.loads(pd.DataFrame(df_start2).to_json(orient='records')))
```

```python
ex_df2.drop('data.Responses.lines.vehiclePositions', inplace=True, axis=1)
```

```python
ex_df2.rename(columns = {'data.time':'time', 'data.Responses.lines.lineId':'lineID','data.Responses.lines.vehiclePositions.directionId':'directionId','data.Responses.lines.vehiclePositions.distanceFromPoint':'distanceFromPoint','data.Responses.lines.vehiclePositions.pointId':'pointId'}, inplace = True)
```

```python
ex_df2
```

```python
from datetime import datetime
```

```python
ex_df2["convert"]=pd.to_datetime(ex_df2["time"],unit='ms')
```

```python
ex_df2["date"]=ex_df2["convert"].dt.date
```

```python
ex_df2['time_new']=ex_df2["convert"].dt.time
```

```python
ex_df2
```

```python
import os

for i,x in ex_df2.groupby('date'):
    directory=str(i)
    folder="Data_of_json2"
    path="E:\MINERVA PROJECT 1"
    path_to_date = os.path.join(path,folder, directory)
    if not os.path.exists(path_to_date):
        os.makedirs(path_to_date) 
    for a,b in x.groupby('lineID'):
        #print("a:",a)
        #print("b:",b)  
        p = os.path.join(path_to_date,"{}.csv".format(a))
        b.to_csv(p, index=False)
        
    
    
    

    
```
