```python
import os

for i,x in ex_df1.groupby('date'):
    directory=str(i)
    folder="All_data_of_json1"
    path="E:\MINERVA PROJECT 1"
    path_to_date = os.path.join(path,folder, directory)
    if not os.path.exists(path_to_date):
        os.makedirs(path_to_date) 
    for a,b in x.groupby('lineID'):
        #print("a:",a)
        #print("b:",b)  
        p = os.path.join(path_to_date,"{}.csv".format(a))
        b.to_csv(p, index=False)
```
